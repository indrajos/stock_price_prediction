```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go
import plotly.offline as py
import plotly.express as px
```

```python
stock_w = pd.read_csv(r'c:\\Users\\SN\\Desktop\\stock_predict\\stock_price_prediction\\BA.csv')
display(stock_w)
```

```python
trace1 = go.Bar(
    x = stock_w['Date'],
    y = stock_w['Adj Close'],
    name='Influenza',
    marker=dict(color='rgba(16, 112, 2, 0.8)',
                line=dict(color='rgb(0,0,0)', width=1.5)),
     
)

data = [trace1]
layout = dict(
    title='Boeing Adj Close',
    xaxis=dict( 
        ticks='outside', ticklen=5, 
        zeroline=True, linewidth=1, linecolor='black'
        )
    )
fig = go.Figure(data=data, layout=layout)            
py.iplot(fig)
```

```python
date_range = stock_w[stock_w["Date"] > "2016-01-01"]
display(date_range)
```

```python
trace1 = go.Scatter(
    x = date_range['Date'],
    y = date_range['Low'],
    mode='lines+markers',
    name='Low',
    marker=dict(color='rgba(16, 112, 2, 0.8)'),
)
trace2 = go.Scatter(
    x = date_range['Date'],
    y = date_range['High'],
    mode='lines+markers',
    name='High',
    marker=dict(color='rgba(80, 26, 80, 0.8)'),
)
data = [trace1, trace2]
layout = dict(
    title='Boeing 2016-2022',
    xaxis=dict( 
        ticks='outside', ticklen=5, 
        zeroline=True, linewidth=1, linecolor='black'
        ))
fig = dict(data=data, layout=layout)            
py.iplot(fig)
```
